```python
# !pip install umap-learn==0.4.6 --no-deps


import pandas as pd
import sklearn
from sklearn.preprocessing import StandardScaler, LabelEncoder

data = pd.read_csv('/kaggle/input/breast-cancer-wisconsin-data/data.csv')

print(data.shape)
data.head()
```

```python
#standardize and create encoding
data = data.drop(columns=['Unnamed: 32']) #no NAN values
x = data.drop(columns=['id', 'diagnosis'])
y = LabelEncoder().fit_transform(data['diagnosis'])
features = x.columns.tolist() #save feature names
x_standardized = StandardScaler().fit_transform(x)
```

```python
#create embeddings
from sklearn.decomposition import PCA
import umap
import seaborn as sns
import matplotlib.pyplot as plt

print("sklearn version:", sklearn.__version__)
print("umap version:", umap.__version__)


seed = 0
pca_2D = PCA(n_components=2, random_state = seed).fit_transform(x_standardized)
reducer = umap.UMAP(n_components=2, random_state = seed)
umap_2D = reducer.fit_transform(x_standardized)

def plot(array, classification, name):
    sns.scatterplot(x=array[:,0], y=array[:,1], hue=classification, palette=['blue', 'red'])
    plt.title(name)
    plt.xlabel('Dim 1'); plt.ylabel('Dim 2')
    plt.legend(title='Diagnosis', labels=['Benign','Malignant'])
    plt.savefig()

plot(pca_2D, data['diagnosis'], "PCA 2D Embeddings")

plot(umap_2D, data['diagnosis'], "UMAP 2D Embeddings")
    
    

```
